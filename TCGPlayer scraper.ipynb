importing all necessary packages for data analysis of the TCG Player Prices Data From TCGPlayer.com

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

```

Getting The URL and setting up the get request for Beautiful Soup to scrape the pages. 

```python
url = 'https://shop.tcgplayer.com/price-guide/flesh-and-blood-tcg'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

Making the scraper go through the dropdown menu for the different sets by iterating through that menu and then looping through the columns and rows of the table in each of the pages available in the drop down menu. this gives me my scraped data. 

```python

#The Url has a dropdown menu I am using the ID of set from the html to select it and move through these pages
dropdown_menu = soup.find('select', id='set')
# If the menu has a table in the set name elemnent I want tot hen iterate through columns and rows. 
if dropdown_menu:

    set_names = [option.text.strip() for option in dropdown_menu.find_all('option')]

data_list = []

for set_name in set_names:
        set_url = f"{url}/{set_name}"  
        set_response = requests.get(set_url)
        set_soup = BeautifulSoup(set_response.content, 'html.parser')

        # Find the table containing the data for the set
        table = set_soup.find('table', class_='priceGuideTable')

        if table:
            
            rows = table.tbody.find_all('tr')

            # Extract and print the data from each row
            for row in rows:
                columns = row.find_all('td')
                data = [column.text.strip() for column in columns]
                print(data)
                print("-------------------------")
        else:
            print("Table not found.")
else:
    print("Dropdown menu not found.")
```

```python
scraped_df = pd.DataFrame(data_list, columns=['Product', 'Rarity', 'Number', 'Market Price', 'Listed Median'])
print(scraped_df)
```
